# heat map to show correlation

```python
df=pd.read_csv('Attendence_report_January (1).csv')
df
```

```python
df.columns
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

df1=df.drop(columns=['Name','Normal Working Hours',' OT Hours ',
'Late Coming Hours',' Early Going Hours '], axis=1)
corr=df1.corr()
sns.heatmap(corr, annot=True, cmap="coolwarm",linewidths=0.5)
plt.title("Correlation Matrix of attendence Dataset")
plt.show()
```

# pair plot the total dataset by hue as present days

```python
sns.pairplot(df,hue="Present Days",diag_kind="hist")
plt.show()
```
